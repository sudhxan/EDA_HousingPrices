```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split


# Assuming 'target' the target column
X = new_df.drop('SalePrice', axis=1)
y = new_df['SalePrice']

encoder = OneHotEncoder(sparse=False, handle_unknown='ignore') #encoding the data
X_encoded = encoder.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.3, random_state=42) #train test split

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

importances = model.feature_importances_
feature_names = encoder.get_feature_names_out(X.columns)
feature_importances = pd.Series(importances, index=feature_names).sort_values(ascending=False)

print("Feature importances:\n", feature_importances.sort_values())
```
